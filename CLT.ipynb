```python
# https://en.wikipedia.org/wiki/Central_limit_theorem

import datetime
import json
import random

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from IPython.core.display import display, HTML

display(HTML("<style>.container { width:80% !important; }</style>"))
```

```python
# sample function
def sample(data, label):
    means = []

    for i in range(1000):
        sample = np.random.choice(data, 50)
        means.append(sample.mean())

    plt.hist(means, 100, label=f'{label}_sample')
    plt.legend()
    plt.show()
    
np.random.seed()
```

```python
# data1
data1 = np.random.normal(0, 5, 10000)

plt.hist(data1, 100)
plt.show()
```

```python
# data2
data2 = np.random.randint(0, 6, 10000)

plt.hist(data2, 100)
plt.show()
```

```python
# sample
sample(data1, 'data1')
sample(data2, 'data2')
```
